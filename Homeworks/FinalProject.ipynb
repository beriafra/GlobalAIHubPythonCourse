```python
quiz = { 1 : {
        "question" : "What is the capital city of Turkey?",
        "answer" : "Ankara"
    },
    2 : {
        "question" : "Who is Harry Potter's best guy friend?",
        "answer" : "Ron Weasley"
    },
    3 : {
        "question" : "How many continents are there in the world?",
        "answer" : "7"
    },
    4 : {
        "question" : "What is the biggest country in the world?",
        "answer" : "Russia"
    },
    5 : {
        "question" : "What is pi  equal to (3.d.p)",
        "answer" : "3.14"
    },
    6 : {
        "question" : "What is the name of PythonCourse lecturer?",
        "answer" : "omer cengiz"
    },
    7 : {
        "question" : "How many days were GlobalAIHub's Python course?",
        "answer" : "5"
    },
    8 : {
        "question" : "What is the greatest animal?",
        "answer" : "White whale"
    },
    9 : {
        "question" : "Which Avengers character is a green giant?",
        "answer" : "Hulk"    
    },
    10 : {
        "question" : "What is the smallest county?",
        "answer" : "Vatican"    
}
       }

def check_ans(question, ans, attempts, score):
    if quiz[question]['answer'].lower() == ans.lower():
        print(f"Correct Answer! \nYour score is {score + 1}!")
        return True
    else:
        print(f"Wrong Answer :( \nYou have {attempts - 1} left! \nTry again...")
        return False

score = 0
for question in quiz:
    attempts = 3
    while attempts > 0:
        print(quiz[question]['question'])
        answer = input("Enter Answer: ")
        check = check_ans(question, answer, attempts, score)
        if check:
            score += 10
            break
        attempts -= 1
        
if score <=50:
    print(f"Your final score is {score}!\n This is unsuccessful \nThanks for playing")
if score >50:
    print(f"Your final score is {score}!\n This is successful  \nThanks for playing")

```
